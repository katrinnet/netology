```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
A = [[-100, -100, 1],
[-100, 100, 1],
[100, 100, 1],
[0, 0, 1],
[100, -100, 1],
[-100, -100, 1]]


A = np.array(A)

x = A[:,0]
y = A[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

```python
A_m = np.eye(3)
A_m = A_m*0.5 # задаем матрицу масштабирования в 0.5
```

```python
A_1 = A@A_m

x = A@A_m[:,0]
y = A@A_m[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

```python
fi = 130*np.pi/180

# задаем матрицу повотора на 130 градусов против часовой стрелки
A_r = [
    [np.cos(fi), np.sin(fi), 1],
    [-np.sin(fi), np.cos(fi), 1],
    [0, 0, 1]
]

A_2 = A_1@A_r

x = A_2[:,0]
y = A_2[:,1]
plt.plot(x, y)
plt.grid()
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

```python
# задаем матрицу отображения относительно y = x
A_s = [
    [0, 1, 0],
    [1, 0, 0],
    [0, 0, 1]
]
A_s

A_3 = A_2@A_s
x = A_3[:,0]
y = A_3[:,1]
plt.plot(x, y)
plt.grid()
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

# Задание 2

x + z = 2
-x + y - 2*z + w = -2
4*x + z - 2*w = 0
-4*x + 4*y + w = 5

```python
A = np.array([[1, 0, 1, 0], [-1, 1, -2, 1], [4, 0, 1, -2], [-4, 4, 0, 1]])
B = np.array([[2], [-2], [0], [5]])
```

```python
np.linalg.solve(A, B)
```
