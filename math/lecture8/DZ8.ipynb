# Задание 1

Монетку подбрасывают 15 раз, найдите вероятность выпадения простого числа орлов. 

Простые числа до 15 = 1, 2, 3, 5, 7, 11, 13
Следовательно, нужно найти вероятности выпадения этого количества орлов из 15 возможных и просуммировать.
Вероятность выпадения любой одной комбинации 1/(2 ** 15).
Количество комбинаций с одним орлом - 15
Количество комбинаций с 2 орлами - сочетание = 15!/(2!*(15-2)!), аналогично для 3, 5, 7, 11, 13.

```python
import math

p_0 = 1/(2**15)
f15 = math.factorial(15)
p_2 = f15/(math.factorial(2)*math.factorial(15-2))
p_3 = f15/(math.factorial(3)*math.factorial(15-3))
p_5 = f15/(math.factorial(5)*math.factorial(15-5))
p_7 = f15/(math.factorial(7)*math.factorial(15-7))
p_11 = f15/(math.factorial(11)*math.factorial(15-11))
p_13 = f15/(math.factorial(13)*math.factorial(15-13))

p = p_0*(15 + p_2 + p_3 + p_5 + p_7 + p_11 + p_13)
p
```

# Задание 2

```python
norm_1 = np.random.normal(loc=1.0, scale=1.0, size=5000)
norm_2 = np.random.normal(loc=-1.0, scale=2.0, size=5000)
norm_3 = np.random.normal(loc=2.0, scale=0.5, size=5000)
norm_4 = np.random.normal(loc=5.0, scale=5.0, size=5000)
norm_5 = np.random.normal(loc=-4.0, scale=0.2, size=5000)

fig, ax = plt.subplots()
fig.set_figheight(20)
fig.set_figwidth(30)
plt.rcParams['font.size'] = '20'
plt.xlabel('x', fontsize=20)
plt.ylabel('density', fontsize=20)

plt.hist(norm_1, bins=50, density=True, alpha = 0.5, label='loc=1.0, scale=1.0')
plt.hist(norm_2, bins=50, density=True, alpha = 0.5, label='loc=-1.0, scale=2.0')
plt.hist(norm_3, bins=50, density=True, alpha = 0.5, label='loc=2.0, scale=0.5')
plt.hist(norm_4, bins=50, density=True, alpha = 0.5, label='loc=5.0, scale=5.0')
plt.hist(norm_5, bins=50, density=True, alpha = 0.5, label='loc=-4.0, scale=0.2')
plt.legend()
```
