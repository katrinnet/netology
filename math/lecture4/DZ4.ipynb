```python
import numpy as np
import matplotlib.pyplot as plt 
import sympy as sym
```

```python
x = sym.Symbol('x')
```

```python
d1 = sym.diff((sym.sin(2*x+1))**5, x, 1)
d1
```

```python
d2 = sym.diff((sym.sin(2*x+1))**5, x, 2)
d2
```

```python
x_ = np.arange(-5, 5.2, 0.1)
y_ = (np.sin(2*x_+1))**5

y_1 = np.diff(y_) / 0.1
y_2 = np.diff(y_1) / 0.1

fig = plt.figure()

fig.set_figheight(20)
fig.set_figwidth(40)

plt.plot(x_[:-2], y_[:-2], label = 'y(x)')
plt.plot(x_[:-2], y_1[:-1], label = "y'(x)")
plt.plot(x_[:-2], y_2, label = 'y"(x)')
plt.legend(loc=2, prop={'size': 30})
plt.xlabel('x')
plt.ylabel('y')
```
