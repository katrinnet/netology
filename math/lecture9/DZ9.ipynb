```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns
```

```python
rv = stats.gamma(1.34)
```

```python
gen_pop = rv.rvs(10000)
gen_pop.mean()
```

```python
sns.histplot(gen_pop, bins = 30, color='green')
```

```python
for n in range(20):
    mean_list = []
    for i in range(1000):
        data = rv.rvs(size=n+1)
        mean_list.append(data.mean())
    sns.histplot(mean_list, color='green')
    plt.title('выборки из '+str(n+1))
    print(np.mean(mean_list))
    plt.show()
```

Начиная с размера выборки = 13 для подсчета среднего, распределения средних обретает нормальный вид.
