```python
import numpy as np
import matplotlib.pyplot as plt 
import sympy as sym
```

```python
x1 = sym.Symbol('x1')
x2 = sym.Symbol('x2')
x3 = sym.Symbol('x3')
```

```python
f = 2*x1**3 - 3*x1*x2 + 2*x2*x3**2 - x2**3
dx1 = sym.diff(f, x1, 1)
dx2 = sym.diff(f, x2, 1)
dx3 = sym.diff(f, x3, 1)
M = {x1: 2, x2: 1, x3: 1}
```

```python
# вектор градиента в точке М
grad_M = np.array([dx1.subs(M), dx2.subs(M), dx3.subs(M)])
grad_M
```

```python
d2x1 = sym.diff(f, x1, 2).subs(M)
d2x2 = sym.diff(f, x2, 2).subs(M)
d2x3 = sym.diff(f, x3, 2).subs(M)
d2x1x2 = sym.diff(f, x1, x2).subs(M)
d2x1x3 = sym.diff(f, x1, x3).subs(M)
d2x2x3 = sym.diff(f, x2, x3).subs(M)

# матрица Гессе в точке М
Hesse = np.array([d2x1, d2x1x2, d2x1x3, d2x1x2, d2x2, d2x2x3, d2x1x3, d2x2x3, d2x3]).reshape(3, 3)
Hesse
```

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
```

```python
fig = plt.figure()
ax = fig.gca(projection = '3d')

X = np.arange(-5, 5, 0.02)
Y = np.arange(-5, 5, 0.02)
X, Y = np.meshgrid(X, Y)
Z = Y**2 - X**2*Y + 2*X*Y
surf = ax.plot_surface(X, Y, Z, cmap = cm.coolwarm, linewidth=0, antialiased=False)

ax.set_zlim(-50.5, 50.0)
ax.zaxis.set_major_locator(LinearLocator(10))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))
fig.colorbar(surf, shrink=0.5, aspect=5)

# ax.azim = 2
ax.dist = 7

plt.show()
```
