# Визуализация

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
_2015 = pd.read_csv('2015.csv')
_2016 = pd.read_csv('2016.csv')
_2017 = pd.read_csv('2017.csv')
_2018 = pd.read_csv('2018.csv')
_2019 = pd.read_csv('2019.csv')

Europe = _2015[_2015['Region'] == 'Western Europe']['Country'].values

_2015.head()
```

```python
bar_2015 = _2015[['Region', 'Trust (Government Corruption)']].groupby('Region').mean()

bar_2015.plot(kind='bar', title='Trust by regoins all over the world', xlabel='Region',
                                                                   ylabel='Trust (Government Corruption)')
```

```python
"""Вывод: в 2015 самый высокий уровень доверия в странах Австралии и Зеландии, средний - в странах Среднего Востока, 
Северной Америки и Зарадной Европы, в остальных регионах показатель довольно низкий"""
```

```python
plot_2015 = _2015[['Happiness Score','Health (Life Expectancy)']].round(1).groupby('Happiness Score')['Health (Life Expectancy)'].mean()

plot_2015.plot(title='Зависимость продолжительности жизни от уровня счастья', xlabel='уровень счастья', 
               ylabel='продолжительность жизни')
```

```python
"""Вывод: в 2015 в целом наблюдается тренд увеличения продолжительности жизни при повышении оценки уровня счастья, 
но это не всегда так, видны значения с низким уровнем счастья 3.0-4.2 и достаточно высокой продолжительностью жизни 0.7"""
```

```python
reg_2015 = _2015[['Region', 'Economy (GDP per Capita)']]

params = {
         'xtick.labelsize':'x-small',
         'ytick.labelsize':'x-small'}
plt.rcParams.update(params)
plt.xticks(rotation=45)
sns.boxplot(x="Region", y="Economy (GDP per Capita)", data=reg_2015)
```

```python
"""Вывод: в данных 2015 ВВП на душу населения по 3м регионам имеются выбросы (например, какие-то страны Западной Европы явно
опережают остальные по этому показателю), самое высокое медианное значение в Северной Америке"""
```

```python
_2015 = _2015[['Country','Happiness Score', 'Economy (GDP per Capita)', 'Health (Life Expectancy)', 'Freedom', 'Generosity']]
_2016 = _2016[['Country','Happiness Score', 'Economy (GDP per Capita)', 'Health (Life Expectancy)', 'Freedom', 'Generosity']]
_2017 = _2017[['Country','Happiness.Score', 'Economy..GDP.per.Capita.', 'Health..Life.Expectancy.', 'Freedom', 'Generosity']]
_2018 = _2018[['Country or region','Score', 'GDP per capita', 'Healthy life expectancy', 'Freedom to make life choices', 'Generosity']]
_2019 = _2019[['Country or region','Score', 'GDP per capita', 'Healthy life expectancy', 'Freedom to make life choices', 'Generosity']]

_2015.rename(columns={'Happiness Score': 'Happiness', 'Economy (GDP per Capita)':'GDP','Health (Life Expectancy)':'Life Exp'}, inplace=True)
_2016.rename(columns={'Happiness Score': 'Happiness', 'Economy (GDP per Capita)':'GDP','Health (Life Expectancy)':'Life Exp'}, inplace=True)
_2017.rename(columns={'Happiness.Score': 'Happiness', 'Economy..GDP.per.Capita.':'GDP','Health..Life.Expectancy.':'Life Exp'}, inplace=True)
_2018.rename(columns={'Country or region': 'Country', 'Score': 'Happiness', 'GDP per capita':'GDP','Healthy life expectancy':'Life Exp', 'Freedom to make life choices':'Freedom'}, inplace=True)
_2019.rename(columns={'Country or region': 'Country', 'Score': 'Happiness', 'GDP per capita':'GDP','Healthy life expectancy':'Life Exp', 'Freedom to make life choices':'Freedom'}, inplace=True)
_2015['year'] = 2015
_2016['year'] = 2016
_2017['year'] = 2017
_2018['year'] = 2018
_2019['year'] = 2019

all_years_happ = pd.concat([_2015, _2016, _2017, _2018, _2019])

all_years_happ.info()
```

```python
params = {
         'xtick.labelsize':'medium',
         'ytick.labelsize':'medium'}
plt.rcParams.update(params)
plt.xticks(rotation=0)

sns.boxplot(x="year", y="GDP", data=all_years_happ)
```

```python
"""Мировые показатели ВВП по годам не особо менялись, наиболее успешным был 2017"""
```

```python
sns.pairplot(all_years_happ[['Happiness', 'Freedom', 'Generosity']])
```

```python
"""Однозачно можно сказать, что большинство людей не щедрые"""
```

```python
Europe_Freedom = all_years_happ[all_years_happ['Country'].isin(Europe)].pivot_table(
                        index='Country', 
                        columns='year', 
                        values='Freedom')

Europe_Freedom
```

```python
sns.heatmap(Europe_Freedom)
```

```python
"""Греция не свободна, Италия почти не свободна, все были свободнее в 2018"""
```

```python
Europe_GDP = all_years_happ[all_years_happ['Country'].isin(Europe)].pivot_table(
                        index='Country', 
                        columns='year', 
                        values='GDP')

Europe_GDP

sns.heatmap(Europe_GDP)
```
