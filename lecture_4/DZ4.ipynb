# Задание 1 и 2 

```python
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}

def doc_number_get():
    doc_number = input('Введите номер документа:')
    return doc_number

def doc_type_get():
    doc_type = input('Введите тип документа:')
    return doc_type

def doc_name_get():
    doc_name = input('Введите владельца документа:')
    return doc_name

def dir_number_get():
    dir_number = input('Введите номер полки:')
    return dir_number

def add_directory():
    for doc in documents:
        for key, value in directories.items():
            if doc['number'] in value:
                doc.setdefault('directory', key)
                
def doc_index_get():
    doc_num = doc_number_get()
    for i, doc in enumerate(documents):
        if doc['number'] == doc_num:
            return i

def doc_dir_get(doc_num):
    for key, value in directories.items():
        if doc_num in value:
            return key
        
def l_command():
    add_directory()
    for doc in documents:
        print(f"№: {doc['number']}, тип: {doc['type']}, владелец: {doc['name']}, полка хранения: {doc['directory']}")

def p_command():
    index_doc = doc_index_get()
    if index_doc:
        print(f"Владелец документа: {documents[index_doc]['name']}")
        return index_doc
    else:
        print(f"Документ не найден в базе")
        
def s_command():
    doc_num = doc_number_get()
    key = doc_dir_get(doc_num)
    if key:
        print(f'Документ хранится на полке: {key}')
    else:
        print(f"Документ не найден в базе")
        
def as_command():
    dir_num = dir_number_get()
    if dir_num in directories.keys():
        print(f'Такая полка уже существует. Текущий перечень полок: {", ".join(directories.keys())}.')
    else:
        directories.setdefault(dir_num,[])
        print(f'Полка добавлена. Текущий перечень полок: {", ".join(directories.keys())}.')
      
def ds_command():
    dir_num = dir_number_get()
    if dir_num in directories.keys() and directories[dir_num] == []:
        del(directories[dir_num])
        print(f'Полка удалена. Текущий перечень полок: {", ".join(directories.keys())}.')
    elif dir_num in directories.keys() and directories[dir_num] != []:
        print(f'На полке есть документы, удалите их перед удалением полки. Текущий перечень полок: {", ".join(directories.keys())}.')
    else:
        print(f'Такой полки не существует. Текущий перечень полок: {", ".join(directories.keys())}.')
    
def ad_command():
    doc_num = doc_number_get()
    doc_type = doc_type_get()
    doc_name = doc_name_get()
    dir_num = dir_number_get()
    
    if dir_num in directories.keys():
        documents.append({'type': doc_type, 'number': doc_num, 'name': doc_name, 'directory': dir_num})
        directories[dir_num].append(doc_num)
        print('Документ добавлен. Текущий список документов:')
        l_command()
    else:
        print('Такой полки не существует. Добавьте полку командой as.')
        print('Текущий список документов:')
        l_command()
        
def d_command():
    index_doc = doc_index_get()
    if index_doc:
            documents.remove(documents[index_doc])
            print('Документ удален.')
            print('Текущий список документов:')
            l_command()
    else:
        print(f"Документ не найден в базе")
        print('Текущий список документов:')
        l_command()
        
def m_command():
    doc_index = doc_index_get()
    dir_num_new = dir_number_get()        
    if dir_num_new not in directories.keys():
        print(f'Такой полки не существует. Текущий перечень полок: {", ".join(directories.keys())}.')
    if doc_index is None:
        print(f"Документ не найден в базе")
        print('Текущий список документов:')
        l_command()
    if doc_index and dir_num_new in directories.keys():
        dir_num = doc_dir_get(documents[doc_index]['number'])
        add_directory()
        documents[doc_index]['directory'] = dir_num_new
        directories[dir_num].remove(documents[doc_index]['number'])
        directories[dir_num_new].append(documents[doc_index]['number'])
        print('Документ перемещен.')
        print('Текущий список документов:')
        l_command()
    
while True:
    command = input('Введите команду:')
    if command == 'q':
        break
    elif command == 'p':
        p_command()
    elif command == 's':
        s_command()
    elif command == 'l':
        l_command()
    elif command == 'as':
        as_command()  
    elif command == 'ds':
        ds_command()
    elif command == 'ad':
        ad_command()
    elif command == 'd':
        d_command()
    elif command == 'm':
        m_command()
        
```
