# Задание 1

```python
import requests
from bs4 import BeautifulSoup
import time
import pandas as pd

KEYWORDS = ['python', 'парсинг']
URL = 'https://habr.com/ru/search'

def get_all_links(url, query, pages):
    all_refs = []

    params = {
    'q': query
    }
    
    for i in range(pages):
        res = requests.get(url+'/page'+str(i+1)+'/', params)
        time.sleep(0.3)
        soup = BeautifulSoup(res.text, 'html.parser')
        posts = soup.find_all('article')
        news_blocks = soup.find_all('article', class_='tm-articles-list__item')
        articles_intro = list(map(lambda x: x.find('div', class_='tm-article-snippet'), news_blocks))
        try:
            """кое-где впилины рекламые посты, у которых class_ другие, их пропускаем"""
            articles_intro_ = list(map(lambda x: x.find('h2', class_='tm-article-snippet__title'), articles_intro))
            a_list = list(map(lambda x: x.find('a').get('href'), articles_intro_))
            all_refs += list(map(lambda x: 'https://habr.com/' + x[1:], a_list))
        except:
            pass
    
    return all_refs

all_links = get_all_links(URL, KEYWORDS, 38)

def get_habr_news(links):
    habr_news = pd.DataFrame()
    for link in all_links:
        soup = BeautifulSoup(requests.get(link).text, 'html.parser')
        time.sleep(0.3)
        date = pd.to_datetime(soup.find('time').get('datetime'), dayfirst=True).date()
        title = soup.find('h1', class_='tm-article-snippet__title').text
        text = soup.find('div', class_='article-formatted-body').text
        row = {'date': date, 'title': title,'link': link, 'text': text}
        habr_news = pd.concat([habr_news, pd.DataFrame([row])])  
    return habr_news

habr_news = get_habr_news(all_links)
habr_news
```
