# Задание 1

```python
import pandas as pd

horse = pd.read_csv('horse_data.csv')
horse.columns = range(0,28)
horse = horse[[0, 1, 3, 4, 5, 6, 10, 22]]

horse.columns = ['surgery', 'age', 'rectal_temp', 'pulse', 'resp_rate', 'temp_of_extremities', 'pain', 'outcome']
horse.replace('?','', inplace=True)
horse['surgery'] = pd.to_numeric(horse['surgery'])
horse['age'] = pd.to_numeric(horse['age'])
horse['rectal_temp'] = pd.to_numeric(horse['rectal_temp'])
horse['pulse'] = pd.to_numeric(horse['pulse'])
horse['resp_rate'] = pd.to_numeric(horse['resp_rate'])
horse['temp_of_extremities'] = pd.to_numeric(horse['temp_of_extremities'])
horse['pain'] = pd.to_numeric(horse['pain'])
horse['outcome'] = pd.to_numeric(horse['outcome'])

horse.head(20)
```

# Задание 2

```python
"""вывод: судя по статистике, в колонках surgery, temp_of_extremities, pain и outcome выбросов нет"""
horse.describe()
```

```python
"""в колонке age выброс 9, встречается 24 раза"""
print(horse[horse['age'] == 9]['age'].count())

horse[horse['age'] == 9]
```

```python
horse.median()
```

```python
"""в колонке rectal_temp выбросов 14"""

q1 = horse['rectal_temp'].quantile(0.25)
q3 = horse['rectal_temp'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
outliers_rec = horse[(horse['rectal_temp'] < lower_bound) | (horse['rectal_temp'] > upper_bound)].sort_values('rectal_temp')
outliers_rec
```

```python
"""в колонке resp_rate выбросов 17, все они значительно выше 75 процентиля, все они с высоким зачением rectal_temp"""

q1 = horse['resp_rate'].quantile(0.25)
q3 = horse['resp_rate'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
outliers_resp = horse[(horse['resp_rate'] < lower_bound) | (horse['resp_rate'] > upper_bound)].sort_values('resp_rate')
outliers_resp
```

```python
"""в колонке pulse выбросов 5, все они значительно выше 75 процентиля, все они с age = 9 и высоким зачением rectal_temp"""

q1 = horse['pulse'].quantile(0.25)
q3 = horse['pulse'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
outliers_pulse = horse[(horse['pulse'] < lower_bound) | (horse['pulse'] > upper_bound)].sort_values('pulse')
outliers_pulse
```

# Задание 3

```python
"""по колонке age пропусков нет, по колонке surgery и outcome по одному пропуску, в колонке pulse 24 пропуска, в остальных около 60"""

horse.count()
```

```python
"""Пропуск surgery заполним =1, таких значений большинство, аналогично для outcome"""

horse['surgery'].fillna(horse['surgery'].mode()[0], inplace=True)
horse['outcome'].fillna(horse['outcome'].mode()[0], inplace=True)
```

```python
"""Есть строки, где пропуски по всем 5 колонкам кроме surgery, age и outcome, их удаление не привело бы к значительному искажению статистики по 3м другим колонкам"""

horse[(horse['temp_of_extremities'].isna()) & (horse['pain'].isna()) & (horse['rectal_temp'].isna()) & (horse['pulse'].isna()) & (horse['resp_rate'].isna())]
horse.drop(horse[(horse['temp_of_extremities'].isna()) & (horse['pain'].isna()) & (horse['rectal_temp'].isna()) & (horse['pulse'].isna()) & (horse['resp_rate'].isna())].index, inplace=True)
horse.describe()
```

```python
"""Когда pain = 2 и surgery = 1, resp_rate около 35.0, когда pain = 2 и surgery = 1, resp_rate около 24.0, заполним этими средними пропуски"""

horse[(horse['surgery'] == 1) & (horse['pain'] == 2)]['resp_rate'].median()
horse[(horse['surgery'] == 2) & (horse['pain'] == 2)]['resp_rate'].median()

horse.loc[(horse['surgery'] == 1) & (horse['pain'] == 2), ['resp_rate']] = horse[['resp_rate']].fillna(horse[(horse['surgery'] == 1) & (horse['pain'] == 2)]['resp_rate'].median(), axis=1)
horse.loc[(horse['surgery'] == 2) & (horse['pain'] == 2), ['resp_rate']] = horse[['resp_rate']].fillna(horse[(horse['surgery'] == 2) & (horse['pain'] == 2)]['resp_rate'].median(), axis=1)
```

```python
horse.describe()
```

```python
"""в строках, где pain = 5, видно, что значения rectal_temp близки к 38 и pulse примерно 89, temp_of_extremities 5, а resp_rate = 30 заполним в этих колонках значения по медианным"""
horse[horse['pain'] == 5].median()
```

```python
rectal_median = horse[horse['pain'] == 5]['rectal_temp'].median()
pulse_median = horse[horse['pain'] == 5]['pulse'].median()
temp_of_extremities_median = horse[horse['pain'] == 5]['temp_of_extremities'].median()
resp_rate_median = horse[horse['pain'] == 5]['resp_rate'].median()

horse.loc[horse['pain'] == 5, ['rectal_temp']] = horse[['rectal_temp']].fillna(rectal_median, axis=1)
horse.loc[horse['pain'] == 5, ['pulse']] = horse[['pulse']].fillna(pulse_median, axis=1)
horse.loc[horse['pain'] == 5, ['temp_of_extremities']] = horse[['temp_of_extremities']].fillna(temp_of_extremities_median, axis=1)
horse.loc[horse['pain'] == 5, ['resp_rate']] = horse[['resp_rate']].fillna(resp_rate_median, axis=1)
```

```python
horse.count()
```

```python
"""Если pulse больше 90, то заметно растет средние rectal_temp и resp_rate, заполним их"""

pulse_90_rectal = horse[horse['pulse'] > 90]['rectal_temp'].median()
pulse_90_resp = horse[horse['pulse'] > 90]['resp_rate'].median()


horse.loc[horse['pulse'] > 90, ['rectal_temp']] = horse[['rectal_temp']].fillna(pulse_90_rectal, axis=1)
horse.loc[horse['pulse'] > 90, ['resp_rate']] = horse[['resp_rate']].fillna(pulse_90_rectal, axis=1)
horse.count()
```

```python
"""оставшееся заполняем модами/медианой"""
horse['pain'].fillna(horse['pain'].mode()[0], inplace=True)
horse['temp_of_extremities'].fillna(horse['temp_of_extremities'].mode()[0], inplace=True)
horse['pulse'].fillna(horse['pulse'].median(), inplace=True)
horse['resp_rate'].fillna(horse['resp_rate'].median(), inplace=True)
horse['rectal_temp'].fillna(horse['rectal_temp'].median(), inplace=True)
horse.describe()
```
